Para el tercer metodo uso los diccionarios de la siguiente manera

```python
#Primero creamos 2 listas, puedo crear cuantas listas quiera entendiendo que las listas son columnas del dataframe
ciudadesColombia=["Bogotá","Medellín","Cali","Barranquilla"]
ciudadesinfo=[11500000,2500000,2200000,1300000]
```

```python
#Le paso el metodo pandas
df3=pd.DataFrame(ciudades)
```
