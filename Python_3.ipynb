```python
#siempre lo primero que debo hacer es importar pandas y numpy
import pandas as pd
import numpy as np
```

```python
#Ahora subimos el archivo desde drive, ya no tenemos que subirlo desde el ordenador
df_estudiantes = pd.read_csv("/content/drive/MyDrive/StudentsPerformance (1) (1).csv")
df_estudiantes
```

```python
#Calculemos el procentaje de participación
df_estudiantes["gender"].value_counts(normalize=True)*100
```

```python
#Que pasa si lo uso en una columna numerica
df_estudiantes["math score"].value_counts()
```

```python
#Ordenemos lo datos por el valor de matematicas en ascendente
df_estudiantes.sort_values("math score")
```

```python
df_estudiantes= df_estudiantes.sort_values("math score")
```

```python
df_estudiantes
```

```python
#IMPORTANTE PARA DESCARGAR DATOS
df_estudiantes.to_csv("matematicas.csv", index=False)
```

```python
#Ahora vamos a finalizar de forma descendente
df_estudiantes.sort_values("math score",ascending=False)
```

```python
#Tambien podemos ordenar los data frame no solo por una columna sino por dos
#Hay que tener en cuenta que organizará por el primer dato y el segunda será dependiente
df_estudiantes.sort_values(["reading score","writing score"], ascending=False)
```

```python
df_estudiantes.to_csv("Organizado por lectura.csv", index=False)
```

```python
# Para guardar definitivamente, sin creer que se debe volver a reescribir utilizar el siguiente metodo
#De esta manera puerdo guardar los datos con el método Inplace
df_estudiantes.sort_values(["reading score","writing score"], ascending=False,
                           inplace=True)
```

```python
df_estudiantes
```

```python
#Para ordenar valores por texto, para eso usaremos la función lambda
#Con ella convertimos primero los valores en minuscula y luego ordenamos
df_estudiantes.sort_values("race/ethnicity", ascending=True,
                           key=lambda col: col.str.lower())
```
